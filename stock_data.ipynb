```python
# 필요한 것들 import
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 파일 읽기
f = open('C:/Users/SSO/Downloads/stock_data.csv')
data = csv.reader(f, delimiter = ',')
header = next(data)
```

```python
# month = 10, ticker = 'cvs', visualization in graph

# 결과 담을 그룹 만들어 주기
result = []
month = []

# range범위 (12):12개 (0,11): 12개
for i in range(12):
    month.append([])
```

```python
# 그룹에 데이터 담기
for row in data:
    if row[0].split('-')[1] == '10' and row[1] == 'CVS':
        result.append(float(row[4]))
    if row[1] == 'NVDA':
        month[int(row[0].split('-')[1])-1].append(float(row[4]))
```

```python
plt.plot(result)
plt.show()
```

```python
plt.boxplot(month)
plt.show()
```
