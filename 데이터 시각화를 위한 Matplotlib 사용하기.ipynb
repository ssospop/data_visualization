```python
import matplotlib
import matplotlib.pyplot as plt
# matplotlib의 하위 패키지인 pyplot
import pandas as pd
import numpy as np
# 데이터 분석 위한 pandas와 numpy
```

```python
# 시리즈 생성
a = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0,100,10))
```

```python
a
```

```python
a.plot()
# 위 Series에 해당하는 line함수 생성
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0), columns = ["A", "B", "C","D"], index = np.arange(0, 100, 10))
```

```python
df
```

```python
df.plot()
```

```python
#특정한 하나의 열 그래프 그리고 싶으면 원하는 열 Series로 추출 후 그리기
df["B"].plot()
```
