```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(6,4))
```

```python
df # 값만을 입력해 DataFrame을 생성했다면 default값인 0부터 시작
```

```python
df.columns = ["A", "B", "C", "D"]
df.index = pd.date_range("20160701", periods = 6)
# pandas가 제공하는 datetime index
```

```python
df.index
```

```python
df
```

```python
df["F"] = [1.0, np.nan, 3.5, 6.1, np.nan, 7.0] #NaN도 인위적으로 값으로 포함시킬 수 있다
```

```python
df
```

```python
df.dropna(how="any") # 행 값 중 NaN이 하나라도 포함되어있으면 그 행을 삭제한다 / how = "all" 이면 모든 행이 NaN인 경우에 삭제
```

```python
df.fillna(value = 5.0) # NaN값에 지정한 값을 추가한다
```

```python
df.isnull() # Null 이 있는 경우(NaN) 일 때 True로 표기
```

```python
df.loc[df.isnull()["F"], :] # F열에서 null값이 있는 행만 추출
```
