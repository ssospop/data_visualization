```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# index_col : 특정 행 index로 설정
f_pd = pd.read_csv('C:/Users/SSO/Downloads/stock_data.csv', encoding = 'cp949', index_col = 0)
f_pd.head()
```

```python
# 행 기준 결측값(NA) 제거
f_pd.dropna(axis = 0)
```

```python
# 새로운 행 데이터 추가
f_pd['relative_return'] = f_pd['return'] - f_pd['snpreturn'] 
f_pd['market_price'] = f_pd['shares'] * f_pd['price']
f_pd
```

```python
# relative_return 값이 0.05가 넘는 행들 추출해 top_relreturn 라는 dataFrame에 저장
top_relreturn = f_pd[(f_pd['relative_return'] > 0.05)]
top_relreturn.shape
```

```python
# 히스토그램 생성 rwidth : 바 간 간격
plt.hist(top_relreturn['relative_return'], histtype='bar', rwidth=0.9)
plt.show()
```

```python
top_marvalue = f_pd[f_pd['market_price'] > 100000000]
plt.scatter(top_marvalue['relative_return'], top_marvalue['market_price'], label = 'skitscat', color = 'r', marker = 'o')
plt.xlabel('Relative Return')
plt.ylabel('Market Value')
plt.show()
```
